```python
import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions
import csv
```

```python
options = PipelineOptions()
p = beam.Pipeline(options=options)
```

```python
from apache_beam.io import ReadFromText
from apache_beam.io import WriteToText
```

```python
def printer(data_item):
    print(data_item[:4])
```

```python
def parse_file(element):
    for line in csv.reader([element], quotechar='"', delimiter=',', quoting=csv.QUOTE_ALL, skipinitialspace=True):
        return line
```

```python
data_from_source = (p
                    | 'ReadMyFile' >> ReadFromText('./lemonade.csv')
                    | 'Parse file' >> beam.Map(parse_file)
                    | 'Print the data' >> beam.ParDo(printer)
                   )
```

```python
# Run the pipeline
result = p.run()

# lets wait until result a available
result.wait_until_finish()

# print the output
# print(output)
```
